#Exploratory data analysis 

Calculating the count for each category in the Gender column and then visualizing it



```python
# plotting a bar chart for Gender and it's count
plt.figure(figsize=(5,5))
ax=sns.countplot(x='Gender',data=df)
plt.title("Gender counts (Male vs Female) ")
for bars in ax.containers:
      ax.bar_label(bars)
```

This graph helps us visually compare the count of gender from the Gender column

Creating a bar chart to compare the relation between Gender and Amount

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gen
```

```python
# plotting a bar chart for gender vs total amount

plt.figure(figsize=(4,4))
plt.title("Amount spent form Buyers (Male vs Female)")

sns.barplot(x= 'Gender', y= 'Amount', data=sales_gen)
```

This graph helps us view the total amount spent based on their gender. As we can see, the total amount spent by F is more then the amount spent by M.

Here is the count for the different age groups, considering the gender

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', color= 'r')
plt.title("Number of people in each Age Group based on Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

This graph helps visualize the count of people based on different age groups. As we can observe, the highest count is of F in the age group 26-35.



```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

```python
# barplot Total Amount vs Age Group
plt.title('Percentage of Amount spent a/c to Age Groups')
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
total = sum(sales_age["Amount"])
for i, v in enumerate(sales_age["Amount"]):
     percentage = v/total *100
     plt.text(i, v, f'{percentage:.1f}%', ha="center", va="bottom")
```

This graph helps visualize the total amount spent by each age group. As we can see, the total amount spent by the age group 36-45 is most as it is also the age group with the most number of people.

```python
plt.figure(figsize=(6,5))
x=df['Age Group'].value_counts().index
y=df['Age Group'].value_counts().values
plt.pie(y, labels=x,startangle =45,autopct='%0.2f%%')
plt.legend(loc=3)
plt.show()
```

```python
# total number of orders from top 10 states

sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by= 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,4)})  
sns.barplot(data= sales_state , x= 'State', y= 'Orders')
plt.title("Number of orders from top 10 states")
plt.xticks(rotation=45)
plt.show()

```

In this graph we can observe the total orders for the top 10 states. We can figure out the which states generate more orders and which states have less orders. This informs us which states to focus on promoting the products.

```python
# total amount/sales  from top 10 states
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(10,4)})  
sns.barplot(data= sales_state , x= 'State', y= 'Amount')
plt.title("Total amount/sales for top 10 states")
plt.xticks(rotation=45)
plt.show()
```

In this graph we can observe the total amount spent by the top 10 states. We can figure out the which states generate more amount and which states generate lesser. This informs us which states to focus on to increase the total amount.
